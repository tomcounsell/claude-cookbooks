# 1. Messages & Basic Chat

Send a single message to Claude, inspect the raw response, then build a minimal `Chat` class that tracks turn history so multi-turn conversations work.

```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic
from rich import print
import json

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
def object_print(object): 
    return print(json.dumps(object.model_dump(), indent=2, default=str))
```

```python
response = client.messages.create(
    model=model,
    max_tokens=256,
    messages=[{"role": "user", "content": "Say hello in one sentence."}],
)

print(response.content[0].text)
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)
```

```python
object_print(message)
```

```python
class Chat:
    def __init__(self):
        # Start with an empty message list
        self.messages = []
    
    def add_user_message(self, text):
        user_message = {
            "role": "user", 
            "content": text
        }
        self.messages.append(user_message)
    
    def add_assistant_message(self, text):
        assistant_message = {
            "role": "assistant", 
            "content": text
        }
        self.messages.append(assistant_message)
    
    def run(self):
        message = client.messages.create(
            model=model,
            max_tokens=1000,
            messages=self.messages,
        )
        text = message.content[0].text
        self.add_assistant_message(text)
        return text
```

```python
chat = Chat()
chat.add_user_message("What rhymes with banana?")
chat.run()
print(chat.messages)
```

```python
chat.add_user_message("you're clever! now write the chorus to my banana song")
text = chat.run()
print(text)
```
