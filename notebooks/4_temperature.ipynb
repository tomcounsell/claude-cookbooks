# 4. Temperature & Personality

Add a `temperature` knob to the chatbot and pass it through to `messages.create`. Subclass with a higher temperature (e.g. `JokeBot`) to dial up creative, less predictable responses.

```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic
from rich import print
import json

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
def object_print(object): 
    return print(json.dumps(object.model_dump(), indent=2, default=str))
```

```python
class ChatBot:
    """
    You are a helpful assistant
    """
    temperature = 0.7
    
    def __init__(self):
        self.messages = list()
        self.system_prompt = self.__doc__.strip()
    
    def add_user_message(self, text):
        user_message = {
            "role": "user", 
            "content": text
        }
        self.messages.append(user_message)
    
    def add_assistant_message(self, text):
        assistant_message = {
            "role": "assistant", 
            "content": text
        }
        self.messages.append(assistant_message)
    
    def run(self):
        if not self.messages:
            intro_text = "How can I help you?"
            print(intro_text)
            
        while True:
            user_input = input("> ")
            if user_input.lower() in ["exit", "stop", "bye"]:
                break
            self.add_user_message(user_input)
            message = client.messages.create(
                model=model,
                max_tokens=1000,
                messages=self.messages,
                system=self.system_prompt,
                temperature=self.temperature,
            )
            
            text = message.content[0].text
            self.add_assistant_message(text)
            print(text)
```

```python
class JokeBot(ChatBot):
    """
    You are a absolute jokester.
    You are incapable of having a serious conversation.
    Everything you say and do is ironic to the extreme.
    """
    temperature = 1
```

```python
chatbot = JokeBot()
chatbot.run()
```
